```python
import pandas as pd
```

```python
vines_df = pd.read_csv('vine_table.csv')
vines_df.head()
```

```python
more_than_20_votes_df = vines_df.loc[(vines_df["total_votes"] >= 20)]
more_than_20_votes_df
```

```python
filtered_df = more_than_20_votes_df.loc[(more_than_20_votes_df["helpful_votes"] / more_than_20_votes_df["total_votes"] >= .5)]
filtered_df
```

```python
paid_review_df = filtered_df.loc[(filtered_df["vine"] == "Y")]
paid_review_df
```

```python
unpaid_review_df = filtered_df.loc[(filtered_df["vine"] == "N")]
unpaid_review_df
```

```python
paid_tot_reviews = len(paid_review_df)
total_paid_reviews = "The Total Paid Reviews:  " + str(paid_tot_reviews)

five_star_paid = len(paid_review_df[paid_review_df["star_rating"] == 5.0])
total_paid_5star = "The Total 5 Star Paid Reviews:  " + str(five_star_paid)

percent_5star_paid = five_star_paid / paid_tot_reviews
percentage_5star_p = "The percentage of paid 5Star Reviews:  " + str(percent_5star_paid)

unpaid_tot_reviews = len(unpaid_review_df)
total_unpaid_reviews = "The Total Unpaid Reviews:  " + str(unpaid_tot_reviews)

five_star_unpaid = len(unpaid_review_df[unpaid_review_df["star_rating"] == 5.0])
total_unpaid_5star = "The Total 5 Star Unpaid Reviews:  " + str(five_star_unpaid)

percent_5_star_unpaid = five_star_unpaid / unpaid_tot_reviews
percentage_5star_up = "The percentage of unpaid 5Star Reviews:  " + str(percent_5_star_unpaid)

print(total_paid_reviews)
print(total_paid_5star)
print(percentage_5star_p)
print("\n")
print(total_unpaid_reviews)
print(total_unpaid_5star)
print(percentage_5star_up)
```
